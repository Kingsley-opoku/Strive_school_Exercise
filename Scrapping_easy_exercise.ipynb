```python
# importing the required libraries 
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# getting target website

page="https://forecast.weather.gov/MapClick.php?x=276&y=148&site=lox&zmx=&zmy=&map_x=276&map_y=148#.Yisn0HrMK00"
```

```python
#pinging the webpage
getting_page=requests.get(page)
getting_page
```

```python
# extracting the required info from the website
extract_data = BeautifulSoup(getting_page.content, "html.parser" )
```

```python
days_info = extract_data.find_all("div" ,class_ ="col-sm-2 forecast-label") # extracting the days 

wheather_description = extract_data.find_all("div" ,class_ ="col-sm-10 forecast-text") # finding the info
```

```python
#extractig the days text from the row data
days_wheather = []
for i in days_info:
    days_wheather.append(i.text)

# exactracting the wheather description text from the raw data
description_wheather = []
for i in wheather_description:
    description_wheather.append(i.text)
```

```python
#Exactracting the temperature from the wheather descriptions

def temperature (text_of_strings):
    list_words= text_of_strings.replace(".", "").split()
    digit_no=[]
    digit_no2=[]
    for j in list_words:
        if j.isdigit():
            digit_no.append(j)
            digit_no2.append(digit_no[0])
    return int(digit_no2[0])

temperatures=[]
for i in description_wheather:

    temp=temperature(i)
    temperatures.append(temp)
    

# converting temperature from Fahrenheit to degree  degrees Celsius (centigrade)

# Celsius= (Fahrenheit-3)(5/9)


def temp_celsius(Fahrenheit):
    celcius=[]
    for i in Fahrenheit:
        ce = ((i - 32) * 5 / 9)
        celcius.append(round(ce,2))
    return celcius

temp_degree=temp_celsius(temperatures)     
   
```

```python
#  definding the time range for the data set
date = (pd.date_range(start= pd.to_datetime("today"), periods=7)       
        .repeat(2)                                                     
        .date                                                       
        )
```

```python
# visualizing the data in pandas dataFrame
df=pd.DataFrame ({'date':date,'days_wheather': days_wheather,  
                    'temp_degree': temp_degree, 
                'description_wheather': description_wheather})
```

```python
# setting the state of the temperature with low or high 
temp_state =(df['description_wheather']                       
                     .str.lower()                  
                     .str.contains('low')               
                     .map({True:'low', False:'high'}))

df['temp_state']=temp_state
```

```python
# rearranging the data column 
data = df.reindex(columns= ['date', 'days_wheather', 'temp_degree','temp_state'] )

data['description_wheather']=description_wheather

data
```
