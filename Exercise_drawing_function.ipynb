```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

## Exercises

1. Plot the absolute value of x and its derivative. Choose an interval to have a nice looking shape.

```python
# definding the range of x
x= np.arange(-5, 5, 0.01)
# definding the absolute value of x
y= abs(x)
#plotting x against y
plt.plot(x,y)
```

```python
#finding the derivate of the absolute value of x
abs_derivative = derivative(abs, x, dx=1e-6)

plt.plot(x, abs_derivative)
```

2. Plot the function $x\log(x)$ and its derivative.

```python
#function for the y=xlog(x)
def xlog(x):
    return x*np.log(x)


x = np.linspace(1, 100, 1000)


plt.plot(x, xlog(x))
```

```python
# finding the derivative of xlog(x) with the step size 10^-6
xlog_derivative=derivative(xlog, x, dx=1e-6)
# ploting x against the derivative of xlog(x)
plt.plot(x, xlog_derivative)
```

Plot the function $e^{x}\log(x)+5$ and its derivative.

```python
def e_xlog(x):
    return np.exp(x)*np.log(x)+5

x = np.linspace(1, 100, 1000)

plt.plot(x, e_xlog(x))
```

```python
dy= derivative(e_xlog, x, dx=1e-6)

plt.plot(x, dy)
```

4.(Harder) Plot the function $\sin(\frac{1}{2}x)+2$ in the interval \[0, $2\pi$\]. Plot in the same graph both the function and the tangent line to the points in which the function has derivative = 0.

```python
def deriv(f,x):
     h = 0.000000001                 #step-size 
     return (f(x+h) - f(x))/h        #definition of derivative

'''Part 2: Plot function with tangent'''
 
def tangent_line(f,x_1,a,b):
     x = np.linspace(a,b,100000)
     y = f(x) 
     y_1 = f(x_1)
     y_tan = deriv(f,x_1) * (x - x_1) + y_1
    
   #plotting
     plt.plot(x,y,'r-')
     plt.plot(x,y_tan,'b-')
     plt.axis([a,b,a,b])
     plt.xlabel('x')     
     plt.ylabel('y')    
     plt.title('Plot of a function with tangent line') 
     plt.show()    

def f2(x):
     return np.sin(0.5*x)+2
 
tangent_line(f2,1,0, np.pi)

```
