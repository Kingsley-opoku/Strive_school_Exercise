```python
import requests
from bs4 import BeautifulSoup
import datetime
import pandas as pd
```

```python
page="https://weather.com/weather/tenday/l/San+Francisco+CA?canonicalCityId=dfdaba8cbe3a4d12a8796e1f7b1ccc7174b4b0a2d5ddb1c8566ae9f154fa638c"
```

```python
getting_page=requests.get(page)
getting_page
```

```python
# extracting the required info from the website
extract_data = BeautifulSoup(getting_page.content, "html.parser" )
```

```python
days_info = extract_data.find_all("h2" ,class_="DetailsSummary--daypartName--2FBp2")

temp= extract_data.find_all('div',class_="DetailsSummary--temperature--1Syw3") 

descriptions=extract_data.find_all('p', class_="DailyContent--narrative--hplRl")

```

```python
# grouping the temperature into low and high
temperatures = []
for i in temp:
    temperatures.append(i.text.replace('Â°', '').split('/'))
    low_temp=[]
    high_temp=[]
    for j in temperatures:
        low_temp.append(int(j[1]))
        high_temp.append(int(j[0]))

# Exactracting the days from the text 
days_wheather = []
for i in days_info:
    days_wheather.append(i.text)

#extracting the wheather descriptions
low_des = []
high_des=[]
for i in descriptions:
    if i.text.find('Low')==-1:
        high_des.append(i.text)
    else:
        low_des.append(i.text)

# function to convert from Fahrenheit to degree celcius
def temp_celsius(Fahrenheit):
    celcius=[]
    for i in Fahrenheit:
        ce = ((i - 32) * 5 / 9)
        celcius.append(round(ce,2))
    return celcius           

low_temp_celcius=temp_celsius(low_temp) #low temperature in degree celcius

high_temp_celcius=temp_celsius(high_temp) # high temperature in degree celcius




date = (pd.date_range(start= pd.to_datetime("today"), periods=15)        #date                                                   
        .date                                                       
        )
```

```python
df=pd.DataFrame ({'date':date,'days_wheather':days_wheather,'low_temp_celcius':low_temp_celcius, 'high_temp_celcius':high_temp_celcius, 'high_des': high_des, 'low_des':low_des} ) 

df
```
